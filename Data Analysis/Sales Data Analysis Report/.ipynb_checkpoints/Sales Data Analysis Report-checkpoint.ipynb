```python
#drop "Row ID" col 'cause it is redundant and change the default index col by "Order ID" col
data = pd.read_excel('Sample - Superstore.xls',index_col='Order ID')
data.drop("Row ID", axis=1, inplace=True)
data.columns
```

```python
#Question 1: How have profit and sales changed over the years?
#firstly, get year from order date
data['Order Year']=data['Order Date'].dt.year
data['Order Year'].head()
```

```python
data['Order Year'].value_counts()
#data report sales from 2014 to 2017
```

```python
data['Discount'].value_counts()
```

```python
change_over_years = data.groupby(['Order Year'])[['Sales','Profit']].agg(['sum']).plot.bar(width = 0.8, fontsize = 14, figsize=(20,10))
plt.xticks(rotation=45)
plt.xlabel('Year', fontsize = 16)
plt.title('Year-Wise Total Sales & Profit Gained', fontsize=20)
plt.legend(['Total Sales', 'Total Profits'], prop={'size': 25})
for bar in change_over_years.patches:
    change_over_years.annotate(format(bar.get_height(), '.2f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')

for i in range(2):
    bar_height = []
    bar_width = []
    for j in range (len(change_over_years.containers[i])):
        bar = change_over_years.containers[i][j]
        bar_height.append(bar.get_height())
        bar_width.append(bar.get_x() + bar.get_width() / 2)
    plt.plot(bar_width, bar_height, marker='o', color='red')
    
plt.show()
# plt.savefig('D:\Machine Learning Project\Internship\Data Analysis\Sales Data Analysis Report\Figs\YearWiseTotalSalesProfit.jpg', format='jpg', dpi=1200)
```

```python
data.groupby(['Order Year'])['Sales'].agg('sum').head()
```

```python
data.groupby(['Order Year'])['Profit'].agg('sum').head()
```

```python
#Question 2: How do profit Rate (Profit/Sales) over years?
#get profit rates by year
profit_rate_over_years = (data.groupby(['Order Year'])['Profit'].agg('sum')/data.groupby(['Order Year'])['Sales'].agg('sum'))*100
profit_rate_over_years.head()
```

```python
sales_change_over_year = data.groupby(['Order Year'])['Sales'].agg("sum")
sales_change_over_year.index
```

```python
profit_rate = profit_rate_over_years.plot( label = 'Profit Margin', color='red', marker='o', fontsize = 14,
                                         xticks = profit_rate_over_years.index,
                                         figsize= (20, 20), linewidth= 4,  markersize=15)

change_over_years = profit_rate.twinx()
change_over_years.bar(sales_change_over_year.index, sales_change_over_year, width = 0.3, alpha = 0.2, label='Total Sales')

for bar in change_over_years.patches:
    change_over_years.annotate(format(bar.get_height(), '.2f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
    
#set legend
bar, labels_1 = change_over_years.get_legend_handles_labels()
line, labels_2 = profit_rate.get_legend_handles_labels()
chart = bar + line
labels = labels_1 + labels_2
profit_rate.legend(chart, labels, loc='upper left', prop={'size': 25})

profit_rate.set_ylabel('Percentage', fontsize = 16, color='red', labelpad=25)
change_over_years.set_ylabel('Total Sales', fontsize = 16, color='blue', labelpad=25)
profit_rate.set_xlabel('Year', fontsize = 24, labelpad=25)
profit_rate.set_xticklabels(profit_rate_over_years.index, rotation=45, fontsize=16)

bar_height = []
bar_width = []
for bar in change_over_years.patches:
    bar_height.append(bar.get_height())
    bar_width.append(bar.get_x() + bar.get_width() / 2)
plt.plot(bar_width, bar_height, marker='o', color='green', markersize=10)

plt.title('Year-Wise Total Sales & Profit Rate', fontsize=24, y=1.02)
plt.grid(True)
# plt.savefig('D:\Machine Learning Project\Internship\Data Analysis\Sales Data Analysis Report\Figs\YearWiseProfitRate.jpg', format='jpg', dpi=1200)
plt.show()
```

```python
#Question 3: How do profit and sales change over months in 2017?
#Get data in 2017
data_2017 = data[data['Order Year']==2017]
data_2017.head()
```

```python
#extract month
data['Order Month'] = data['Order Date'].dt.month
data['Order Month'].head()
```

```python
change_over_month_2017 = data[data['Order Year']==2017].groupby(['Order Month'])[['Sales','Profit']].agg(['sum'])
change_over_month_2017.columns = change_over_month_2017.columns.droplevel(1)
change_over_month_2017.rename(columns = {'Sales':'Total Sales', 'Profit':'Total Profits'}, inplace = True)
change_over_month_2017.index[1]
```

```python
profit_rate_over_months = (data[data['Order Year']==2017].groupby(['Order Month'])['Profit'].agg('sum')/data[data['Order Year']==2017].groupby(['Order Month'])['Sales'].agg('sum'))*100
profit_rate_over_months.index
```

```python
data_2017 = data[data['Order Year']==2017]
data_2017['Loss_Amount'] = 0.0
for i in range(data_2017.shape[0]):
    data_2017['Loss_Amount'].values[i] = (data_2017['Sales'][i]/(1-data_2017['Discount'][i])) - data_2017['Sales'][i]
loss_over_months = data_2017.groupby(['Order Month'])['Loss_Amount'].agg('sum')
loss_over_months.head(12)
```

```python
profit_rate = profit_rate_over_months.plot(xticks = profit_rate_over_months.index, figsize=(30,15), 
                                           label = 'Profit Margin', color='red', marker='o', linewidth= 4,  markersize=15,
                                          fontsize = 24)

ax2 = profit_rate.twiny()
ax3 = ax2.twinx()
ax4 = profit_rate.twiny()
ax5 = ax4.twinx()

change_over_month = change_over_month_2017.plot.bar(stacked= False, ax = ax3,width= 0.8, fontsize = 24, secondary_y = True,
                                                    alpha = 0.3, legend = '')
rspine = ax5.spines['right']
rspine.set_position(('axes', 1.15))
plt.subplots_adjust(right=0.7)

loss_over_months_line = loss_over_months.plot(label = 'Loss Amount', color='m', marker='x', linewidth= 4,  markersize=15,
                                              ax = ax5, fontsize = 24)
loss_over_months_line.set_ylabel("Loss Amount", color = 'm', size = 24)
ax2.set_xticks([])
ax4.set_xticks([])
profit_rate.set_xlabel("Month", fontsize = 30, labelpad = 20)
profit_rate.set_xticklabels(profit_rate_over_months.index ,rotation=45, fontsize = 20)
profit_rate.set_ylabel("Percentage", color = 'red', size = 24)

# set legend
bar, labels_1 = change_over_month.get_legend_handles_labels()
line, labels_2 = profit_rate.get_legend_handles_labels()
line2, labels_3 = loss_over_months_line.get_legend_handles_labels()
chart = bar + line + line2
labels = labels_1 + labels_2 + labels_3
plt.legend(chart, labels, loc='upper left', prop={'size': 25})

for bar in change_over_month.patches:
    change_over_month.annotate(format(bar.get_height(), '.2f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=13, xytext=(0, 20),
                   textcoords='offset points')
for i in range(2):
    bar_height = []
    bar_width = []
    for j in range (len(change_over_month.containers[i])):
        bar = change_over_month.containers[i][j]
        bar_height.append(bar.get_height())
        bar_width.append(bar.get_x() + bar.get_width() / 2)
    plt.plot(bar_width, bar_height, marker='o', color='green')
    
ymax = max(loss_over_months)
xpos = loss_over_months[loss_over_months==ymax].index.values.astype(float)
loss_over_months_line.annotate(ymax, xy=(xpos, ymax), fontsize = 14,
            arrowprops=dict(facecolor='red')
            )

plt.title('Monthly Sales and Profit in 2017', fontsize = 30, y = 1.01)    
# plt.savefig('D:\Machine Learning Project\Internship\Data Analysis\Sales Data Analysis Report\Figs\MonthlySalesandProfitin2017withLoss.jpg', format='jpg', dpi=1200)
plt.show()
```

```python
#Question 8: Which sub-category are usually sold together?
grouped_item = data.sort_values(['Order ID', 'Sub-Category']).groupby('Order ID')['Sub-Category'].unique()
grouped_item.reset_index(drop=True, inplace=True)
grouped_item.size
```

```python
for i in range(grouped_item.size):
    if grouped_item[i].size == 1:
        grouped_item = grouped_item.drop(i)        
```

```python
grouped_item.size
```

```python
#get combo 
subsets = grouped_item.apply(lambda x: set(x)).tolist()
Count = [sum(s2.issubset(s1) for s1 in subsets) for s2 in subsets]
len(Count)
```

```python
combo = grouped_item.apply(lambda x:','.join(x))
combo = combo.reset_index()
combo.head(15)
```

```python
combo['Count'] = Count
```

```python
combo.drop('index', axis=1, inplace=True)
combo.head(15)
```

```python
combo.shape
```

```python
combo.drop_duplicates(subset ='Sub-Category',
                     keep = False, inplace = True)
```

```python
combo.shape
```

```python
combo = combo.sort_values(by='Count', ascending=False).head(10)
count = combo['Count'].squeeze()
combo
```

```python
explode = (0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#9999FF', '#FF99CC', '#99FFCC', '#99FFFF', '#FF99FF', '#ccff99'] 
best_combo_chart = count.plot.pie(labels=None, figsize = (20,20), autopct='%1.1f%%', startangle = 90, explode = explode, colors = colors, fontsize = 24)

best_combo_chart.legend(labels=combo['Sub-Category'], loc= 'upper left',prop={'size': 20})

plt.tight_layout()
plt.title('Top 10 Best-Selling Combo, 2014-2017', fontsize = 24, y = 0)
plt.ylabel(None)

#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
best_combo_chart.axis('equal')  
# plt.savefig('D:\Machine Learning Project\Internship\Data Analysis\Sales Data Analysis Report\Figs\BestSellCombo.jpg', format='jpg', dpi=1200)
plt.show()
```
