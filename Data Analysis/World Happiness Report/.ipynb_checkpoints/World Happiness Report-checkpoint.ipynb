```python
#Question 1: Factors difference between 2015 and 2019?

# First we group the data by year and average the factors
grouped = happinessData.groupby(by = 'Year')[['Happiness_Score', 'GDP',
       'Social support', 'Health',
       'Freedom', 'Trust', 'Generosity']].mean().reset_index()

# Now we reconstruct the df by using melt() function
grouped = pd.melt(frame = grouped, id_vars='Year', value_vars=['Happiness_Score', 'GDP',
       'Social support', 'Health',
       'Freedom', 'Trust', 'Generosity'], var_name='Factor', value_name='Avg_value')

grouped.tail()
```

```python
factor_change_data = grouped[grouped['Factor'] != 'Happiness_Score']
factor_change_data.set_index('Avg_value', inplace = True)
factor_change_data=factor_change_data.groupby(['Factor']).groups
factor_change_df = pd.DataFrame(factor_change_data).T
factor_change_df
```

```python
my_palette = ["#7A92FF", "#FF7AEF", "#B77AFF", "#A9FF7A", "#FFB27A", "#FF7A7A",
             "#7AFEFF", "#D57AFF", "#FFDF7A", "#D3FF7A"]

factor_change = factor_change_df.plot.bar(figsize = (30, 15), color = my_palette[1:], width = 0.8)

plt.title("Difference in Factors - Then and Now - ", fontsize = 30, y = 1.01)
plt.xlabel("Factor", fontsize = 30)
plt.ylabel("Average Score", fontsize = 30)
plt.xticks(fontsize = 20, rotation = 0)
plt.yticks(fontsize = 20)
plt.legend(['2015', '2016', '2017', '2018', '2019'], prop={'size' : 25})
# plt.savefig('D:\Machine Learning Project\Internship\Data Analysis\World Happiness Report\Figs\DiffInFactors.jpg', format='jpg', dpi=1200)
plt.show()
```

```python
#Question 2: Which are the happiest countries in 2019?
country_score_avg = happinessData[happinessData['Year']==2019].groupby(by = ['Country'])['Happiness_Score'].mean().reset_index()
country_score_avg.set_index('Country', inplace = True)
country_score_avg_df = country_score_avg.sort_values(by = 'Happiness_Score', ascending = False).head(10)
country_score_avg_df = country_score_avg_df.sort_values(by = 'Happiness_Score')
country_score_avg_df
```

```python
country_score = country_score_avg_df.plot.barh(figsize = (30, 15))

plt.title("Top 10 Happiest Countries in 2019", fontsize = 30, y = 1.01)
plt.xlabel("Happiness Score", fontsize = 30)
plt.ylabel("Country", fontsize = 30)
plt.xticks(fontsize = 20, rotation = 45)
plt.yticks(fontsize = 20)

i = 0
for bar in country_score.patches:
    bar.set_color(my_palette[i])
    i+=1
    
plt.legend([])
# plt.savefig('D:\Machine Learning Project\Internship\Data Analysis\World Happiness Report\Figs\Top10HappiestCountries2019.jpg', format='jpg', dpi=1200)
plt.show()
```

```python
#Question 8: Which have strong positive impact on happiness score?
happiness_relationship = happinessData.corr()
plt.figure(figsize = (20,20))
sns.set(font_scale=2)
happiness_relationship_map = sns.heatmap(happiness_relationship, annot=True, linewidths=.5)
happiness_relationship_map.set_yticklabels(happiness_relationship.index, rotation=0)
# plt.savefig('D:\Machine Learning Project\Internship\Data Analysis\World Happiness Report\Figs\FactorHeatmap.jpg', format='jpg', dpi=1200)
plt.show()
```
